```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
import sklearn
```

```python

bank_note_data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/training_set_label.csv" )
bank_note_data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_cols=['VWTI','SWTI','CWTI','EI']
bank_note_data[numerical_cols] = scaler.fit_transform(bank_note_data[numerical_cols])
```

```python
model=Sequential()
model.add(Dense(60,activation='relu',input_dim=4))
model.add(Dense(35,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X=bank_note_data.drop(columns='Class')
y=bank_note_data['Class']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
pred=model.predict(X_test)
```

```python
pred= np.where(pred > 0.5, 1,0)
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/testing_set_label.csv')
```

```python
test_data
```

```python
test_data[numerical_cols] = scaler.fit_transform(test_data[numerical_cols])
pred=model.predict(test_data)
```

```python
pred= np.where(pred > 0.5, 1,0)
```

```python
res = pd.DataFrame(pred)
res.index = test_data.index
res.columns = ['prediction']
res.to_csv("assignment1.csv", index=False)
```
